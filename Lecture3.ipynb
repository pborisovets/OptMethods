Kaczmarz method and RKM for equiangled lines in R2

```python
import numpy as np


n = 10 #number of lines
e = [10, 10]
er = [10, 10]

def step_Kacz(xk, theta):
    '''theta in radians'''
    ai = [np.cos(theta), np.sin(theta)]
    xk1 = xk - np.dot(np.dot(ai,xk), ai)
    return xk1

for i in range(100):
    e = step_Kacz(e, i*np.pi/n)
    print("E: ", e, np.linalg.norm(e))
    er = step_Kacz(er, np.random.randint(0,10)*np.pi/10)
    print("Er: ", er, np.linalg.norm(er))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
scale = 10
e = [10, 10]

theta = [i*np.pi/n for i in range(n)]
for t in theta:
    plt.plot([scale*np.cos(t), -scale*np.cos(t)],[scale*np.sin(t), -scale*np.sin(t)])

for i in range(11):
    e1 = step_Kacz(e, i*np.pi/n)
    plt.plot([e[0],e1[0]], [e[1],e1[1]])
    e=e1
plt.axis("equal")
```
